# How Many Classes Occur in a File?

```python
import pandas as pd
import json
import itertools
import collections
import re
import arrow
rcParams['figure.figsize'] = (15, 3)
pd.set_option('display.mpl_style', 'default')
```

```python
log = json.load(open('ipython-log.json'))
print len(log)
```

```python
rows = list(itertools.chain.from_iterable(itertools.chain.from_iterable([
 [
  [
   [fname, cls, commit['commit'], arrow.arrow(commit['committer']['date']).datetime]
   for cls in classes
   ]
  for fname, classes in
  commit['classes'].items()]
 for commit in log
 ])))
df = pd.DataFrame(rows, columns=['fname', 'class', 'commit', 'date'])
df.index = df['date']
df.head(n=2)
```

## OoOoOOoooh look at this pretty graph

```python
fs = {row[0] for row in rows if re.match(r'^IPython/parallel.*\.py$', row[0])}
rcParams['figure.figsize'] = (15, 10)
for f in fs:
    dff = df[df.fname == f].groupby('date').aggregate(len)
    dff.plot(y='fname', legend=True, logy=True)
```
